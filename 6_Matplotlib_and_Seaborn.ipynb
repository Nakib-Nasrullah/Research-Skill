```python
# Simple line plot: First 10 math scores
plt.plot(students['math score'].head(10)) # X-axis auto (row numbers)
plt.title("Math Scores of first 10 students") #Title
plt.xlabel("Student Number") #X label
plt.ylabel('Math Score') # y lable
plt.show() # Show plot
```

```python
# First 10 math scores
scores = students['math score'].head(10)

# Create x-axis as 1 to 10
x = range(1, 11)  # or list(range(1, 11))

plt.plot(x, scores, marker='o')
plt.title("Math Score of First 10 students")
plt.xlabel("Student Number")
plt.ylabel("Math score")
plt.show()
```

# Bar **Chart**

```python
# Count gender
gender_count = students['gender'].value_counts()

# Simple bar chart
plt.bar(['Female', 'Male'], gender_count.values)
plt.title("Gender Count")
plt.show()
```

```python
# Bar chart with color
plt.bar(['Female', 'Male'], gender_count.values, color=['green', 'blue'])
plt.ylabel('Count')
plt.title("Students by Gennder")
plt.show()
```

```python
# Group by gender, average math score
avg_math = students.groupby('gender')['math score'].mean()

# Bar plot
plt.grid(True, which='both', linestyle="--", linewidth=0.7, alpha= 0.7) # import for alltime.
plt.bar(avg_math.index, avg_math.values, color=['lightblue', 'salmon'])
plt.title("Average math score by Gender")
plt.xlabel('Gender')
plt.ylabel('Average MAth Score')
plt.show()
```

# Scatter **Plot**

```python
# Math vs Reading
plt.scatter(students['math score'], students['reading score'])
plt.grid(True)
plt.xlabel('Math')
plt.ylabel('Reading')
plt.title('Math vs Reading')
plt.show()
```

```python
# Math vs Reading
plt.scatter(students['math score'], students['reading score'])
plt.grid(True)
plt.xlabel('Math')
plt.ylabel('Reading')
plt.title('Math vs Reading')
plt.show()
```

# ***Histogram***

```python
# Test preparation
test_prep = students['test preparation course'].value_counts()

plt.pie(test_prep.values, labels=['None', 'Completed'])
plt.title('Test Preparation')
plt.show()
```

```python
# with percentages
plt.pie(test_prep.values, labels = ['None', 'Completed'], autopct='%1.1f%%')
plt.title('Test Prep Status')
plt.show()
```

```python
education_counts = students['parental level of education'].value_counts()
plt.pie(education_counts, labels=education_counts.index, autopct='%1.2f%%')
plt.title('Percental Education Levels')
plt.show()
```

# Multiple **Plots**

```python
# Create 2X2 grid
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# plot 1: Histogram
axes[0,0].hist(students['math score'], bins = 15)
axes[0,0].set_title('Math Score')

# Plot 2: Bar chart
students['gender'].value_counts().plot(kind='bar', ax=axes[0,1])
axes[0,1].set_title('Gender')

# Plot 3: Scatter
axes[1, 0].scatter(students['math score'][:50], students['reading score'][:50])
axes[1, 0].set_title('Math vs Reading')

# Plot 4: Line
axes[1, 1].plot(students['writing score'][:50])
axes[1, 1].set_title('Writing Scores')

plt.tight_layout()
plt.show()
```

# ***Seaborn***

```python
sns.histplot(students['math score'])
plt.title('Math Score with Seaborn')
plt.show()

# With KDE curve
sns.histplot(students['math score'], kde=True)
plt.show()
```

# Box ***Plot***

```python
# GEnder vs Math Score
sns.boxplot(data=students, x ='gender', y ='math score')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.8)
plt.title('Math vs Gender')
plt.show()
```

```python
# Lunch type effect
sns.boxplot(data=students, x ='lunch', y='math score')
plt.grid(True, which='both', linestyle='--', linewidth=0.8, alpha=0.7)
plt.title('Math by Lunch Type')
plt.show()
```

# *Count Plot*

```python
# Count by race/ehnicity
sns.countplot(data=students, x='race/ethnicity')
plt.grid(True, which="both", linestyle='--', linewidth=0.7, alpha=0.7)
plt.title('Race/Ethnicity Count')
plt.show()
```

```python
# Count test prep
sns.countplot(data=students, x='test preparation course')
plt.grid(True, which="both", linestyle='--', linewidth=0.7, alpha=0.7)
plt.title('Test prep count')
plt.show()
```

# Bar Plot with Average


```python
# Average math score by gender
sns.barplot(data=students, x='gender', y='math score')
plt.grid(True, which="both", linestyle='--', linewidth=0.7, alpha=0.7)
plt.title("Average Math Score by Gender")
plt.show()
```

```python
# By parental Education
sns.barplot(data=students, x ='parental level of education', y='math score')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
# Math score by gender
sns.kdeplot(data=students, x ='math score', hue='gender') #Kernel Density Estimation (KDE)plot
plt.grid(True, which="both", linestyle='--', linewidth=0.7, alpha=0.7)
plt.title('Math score by Gender')
plt.show()
```

```python
# Count by race/ethnicity
sns.countplot(data=students, x='race/ethnicity')
plt.title('Race/Ethnicity Count')
plt.show()
```
