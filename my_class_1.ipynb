# Data **Structure**

```python
#Creating a list
# In set there are no access a duplicate value but in list tauple and dictionaray thre are accsee to put the duplicate value.
fruits=["apple","banana", "orange", "mango", "pineapple"]
#Accessing element
print(fruits[0])
print(fruits[2])
print(fruits[-1])
# Slicing a lists
print(fruits[1:4])
fruits[1] = "grape" # Changing an element
print(fruits)
```

```python
#Adding Elements in List
fruits = ["Apple", "banana"]
fruits.append("orange") # Add to an element in the end ["Apple", "banana", "orange"]
print(fruits)
fruits.insert(1, "mango") # Insert at position
print(fruits)
#In tauple you coldn't change the value or update the value.
# In dectionary there are a kee and a value for the kee.

```

```python
# Remove Element in list
fruits = ["apple", "mango", "banana", "orange", "Dove"]
fruits.remove("Dove")
print(fruits)
popped = fruits.pop() #Remove & return last item.
print(popped) # orange
print(fruits)

```

```python
#organiging list
fruits = ["orange", "apple", "mango", "banana"]
fruits.sort() #Alphabatic order
print(fruits)
fruits.reverse()
print(fruits)
```
