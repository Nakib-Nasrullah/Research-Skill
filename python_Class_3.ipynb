```python
#Basic if-else
age = int(input("Enter your age: "))
if age >=18:
  print("You are adult.")
else:
  print("You are a minor")
```

```python
#if-elif-else (muliple conditions)
score = int(input("Enter the score: "))
if score >=90:
  grade = "A+"
elif score >=80:
  grade = "A"
elif score >=70:
  grade = "B"
elif score >=60:
  grade = "C"
else:
  grade = "F"
print(f"Your grade is {grade}")
```

```python
print(f"Your grade is {grade}")
```

```python
#nested if statements
temperature = 32
humidity = 80
if temperature > 30:
  if humidity > 70:
    print("Hot and humid")
  else:
    print("Hot but dry")
else:
  if humidity > 60:
    print("cool but humid")
  else:
    print("cool and dry")
```

```python
if temperature > 30 and humidity > 70:
  print("Hot and humid")
elif temperature > 30 and humidity <= 70:
  print("Hot but dry")
elif temperature <= 30 and humidity > 70:
  print("cool but humid")
else:
    print("Cool and dry")
```

```python
#other logical operator
has_id = True
is_holiday = False
if has_id or is_holiday:
  print("Can enter")
if not is_holiday:
  print("Work day")
```
