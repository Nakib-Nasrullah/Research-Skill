```python
#if-elif-else (muliple conditions)
score = int(input("Enter the score: "))
if score >=90:
  grade = "A+"
elif score >=80:
  grade = "A"
elif score >=70:
  grade = "B"
elif score >=60:
  grade = "C"
else:
  grade = "F"
print(f"Your grade is {grade}")
```

```python
#nested if statements
temperature = 32
humidity = 80
if temperature > 30:
  if humidity > 70:
    print("Hot and humid")
  else:
    print("Hot but dry")
else:
  if humidity > 60:
    print("cool but humid")
  else:
    print("cool and dry")
```

```python
if temperature > 30 and humidity > 70:
  print("Hot and humid")
elif temperature > 30 and humidity <= 70:
  print("Hot but dry")
elif temperature <= 30 and humidity > 70:
  print("cool but humid")
else:
    print("Cool and dry")
```


Loop



```python
#looping through a list
fruits = ["apple", "banana", "orange"]
for i in fruits:
  print(f"I like {i}")
print("\n")
```

```python
#Using range() function
for i in range(10):
  print(i)
print("\n")

for i in range(2,10):
  print(i)
print("\n")
```

```python
#Continue until a condition is met
count = 0
while count <5:
  print(count)
  count += 1 #Don't forgate to increment !!
print("\n")
```

```python
#with a break statement -
number = 0
while True:
  number +=1
  if number > 10:
    break
  print(number)
```

```python
#Continue statemnet - skip to next iteration
for num in range(10):
  if num % 2 == 0:
    continue  #skep even number
  print(num)
print("\n")
```

```python
#Break statement - exit the loop
for num in range(100):
  if num > 10:
    break
  print(num)
```

```python
#Multiplication table (1-5)
for i in range(1, 6):
  for j in range(1, 6):
    product = i * j
    print(f"{i} X {j} = {product}")
  print("-------")
```

```python
#pattern printing
for i in range(1, 10):
  print("*" *i)
```

```python
#Function with return value
def add_numbers(a,b):
  sum = a + b
  return sum
result = add_numbers(5, 7)
print(f"The sum is {result}")
```

```python
def multi_numbers(x,y):
  multip = x*y
  return multip
result = multi_numbers(5,6)
print(f"The multiplication is {result}")
```

```python
def descreption_person(name, age, city):
  print(f"{name} is {age} years old from {city}")
descreption_person("Nakib", "27", "Pirojpur")
```

```python
def calculator_status(numbers):
  total = sum(numbers)
  average = total/len(numbers)
  maximum = max(numbers)
  minimum = min(numbers)
  return total, average, maximum, minimum
stats = calculator_status([5,10,20,30])
print(stats)
```

```python
total = 0 # Global variable
def add_to_total(value):
  global total
  total += value
  return total
print(add_to_total(5))
print(add_to_total(20))
```
