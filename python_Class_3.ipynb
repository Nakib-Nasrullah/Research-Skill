```python
#Basic if-else
age = int(input("Enter your age: "))
if age >=18:
  print("You are adult.")
else:
  print("You are a minor")
```

```python
#if-elif-else (muliple conditions)
score = int(input("Enter the score: "))
if score >=90:
  grade = "A+"
elif score >=80:
  grade = "A"
elif score >=70:
  grade = "B"
elif score >=60:
  grade = "C"
else:
  grade = "F"
print("Grade is : ",grade)
```
