# ***Train-Test Split***

```python
# train (70%) validation (10%) test(20%)

#Separate input (X) and output (y)
X = df.drop('Outcome', axis=1) # Everything except Outcome
y = df['Outcome']              # Only Outcome

print("X has all patiens information (8 colums)")
print("y has diabetes yes/no (1 column)")

# Split: 80% for training, 20% for testing

#Train test split function is very importent for machine learning

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size = 0.2,     # 20% for test
    random_state = 20   # Same result every time # random_state mens in every time same data go to train and same data go to test.
)

print(f"\n We will train with: {len(X_train)} patients")
print(f"We will test with: {len(X_test)} patients")
```

```python
# Problem: Some number are big, some are small
print("Before scaling - different ranges:")
print(f"Age: smallest={X_train['Age'].min()}, biggest={X_train['Age'].max()}")
print(f"Insulin: smallesr={X_train['Insulin'].min()}, biggest={X_train['Insulin'].max()}")
print("\nAge goes from 21 to 81 (small range)")
print("Insulin goes from 0 to 846 (big range!)")
print("This confuses the model!")
```

# *StanderdScaler*

```python
# Creatae scaler
scaler = StandardScaler()

# Learn from training data and scal it
X_train_scaled = scaler.fit_transform(X_train) #fit_trannsform(X_train) are applyed only for X_train data

# Scale test data (just trannsform, don't fit)
X_test_scaled = scaler.transform(X_test) # transform() are applyed for X_test data


print("\nAfter scaling: ")
print("All features now have similar range! ")
print(f"Example - first patient's age before scaling: {X_train.values[0][7]}")
print(f"Example -first patient's age after scaling: {X_train_scaled[0][7]:.2f}")
```

# **MinMaxScaler (Min-Max normalization)**

```python
from sklearn.preprocessing import MinMaxScaler

#Create MinMaxScaler
minmax_scaler = MinMaxScaler()

# Learn from trainug data and scal it
X_train_minmax = minmax_scaler.fit_transform(X_train)

# Scal test data
X_test_minmax = minmax_scaler.transform(X_test)

print("\nAfter MInMaxScaler:")
print("All fearures now between 0 and 1!")
print(f"Example - first patient's age before scaling: {X_train.values[0][7]}")
print(f"Example - first patient's age after scaling: {X_train_minmax[0][7]:.2f}")

# Let's veryfy MinMaxScaler worked
print("\nChecking MinMaxScaler:")
print(f" Minimum value: {X_train_minmax.min():.2f} (should be 0)")
print(f" Minimum value: {X_train_minmax.max():.2f} (should be 1)")
```

# **Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Let's create a sample datasetwith text  data
sample_data = pd.DataFrame({
    'age': [25,30,35,40],
    'gender': ['male', 'female', 'male', 'female'],
    'smoker': ['yes', 'no', 'yes', 'no'],
    'disease': [1, 0, 1, 0]
})

# For yes/no or male/female
label_encoder = LabelEncoder()

# Convert 'smoker' column
sample_data['moker_encoded'] = label_encoder.fit_transform(sample_data['smoker'])
print("\nAfter Lable Encoding 'smoker':")
print(sample_data[['smoker', 'moker_encoded']])
# no = 0, yes = 1

# Connvert 'gender' column
sample_data['gender_encoded'] = label_encoder.fit_transform(sample_data['gender'])
print("\nAfter Lable Encoding 'gender':")
print(sample_data[['gender', 'gender_encoded']])
```

# **One_Hot Encoding**

```python
# Let's add a column with multiple catagories
sample_data['city'] = ['Delhi', 'Mumbai', 'Delhi', 'Kolkata']

# On-Hot Encoding using pandas
city_encoded = pd.get_dummies(sample_data['city'], prefix='city')
print("\nnOne-Hot Encodiing for 'city':")
print(city_encoded)
# Create separate columns for each city

# Add tooriginal data
sample_data =pd.concat([sample_data, city_encoded], axis=1)
print("\nFinal data with all enncodings:")
print(sample_data)
```

# **Train a Simple Model**

```python
from sklearn.linear_model import LogisticRegression
# Create a model
model = LogisticRegression(max_iter=1000)

# Train the model
model.fit(X_train_scaled, y_train)
print("Model trained!!")

# Check accuricy on training data
train_score = model.score(X_train_scaled, y_train)
print(f"Training accuricy:{train_score:.1%}")

# Check accuricy on test data
test_score = model.score(X_test_scaled, y_test)
print(f"Test accuricy: {test_score:.1%}")
```

# **Cross-Validation**

```python
from sklearn.model_selection import cross_val_score

# Do 5-fold Cross-validation
scores = cross_val_score(
    model,             # Our model
    X_train_scaled,    # Training data
    y_train,           # Training lables
    cv = 5             # 5 mini-tests
)

print("5 mini-tesr scores:")
for i in range(5):
  print(f" Test {i+1}: {scores[i]:.1%}")

print(f"\nAverage: {scores.mean():.1%}")
print(f"This means our model is {scores.mean():.1%} accurate!!")
```
